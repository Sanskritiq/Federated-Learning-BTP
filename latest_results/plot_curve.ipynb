```python
paths = [
    'fedavg/mnist/mnist_acc_metrics.csv',
    'fedavgm/mnist/mnist_acc_metrics.csv',
    'fedprox/mnist/mnist_acc_metrics.csv',
    'moon/mnist/mnist_acc_metrics.csv',
    'fedavg/cifar10/cifar10_acc_metrics.csv',
    'fedavgm/cifar10/cifar10_acc_metrics.csv',
    'fedprox/cifar10/cifar10_acc_metrics.csv',
    'moon/cifar10/cifar10_acc_metrics.csv',
]
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv(paths[0])
print(df.columns)
print(df.head())
```

```python
for path in paths[4:]:
    name, dataset, _ = path.split('/')
    df = pd.read_csv(path)
    acc_before = df['accuracy_test_before'].values
    acc_after = df['accuracy_test_after'].values
    epochs = df['epoch'].values
    
    fig = plt.figure()
    plt.plot(epochs, acc_before, label='Before')
    plt.plot(epochs, acc_after, label='After')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.title(f'{name}_{dataset}') 
    plt.legend() 
    fig.savefig(f'{name}_{dataset}.png')  
    
```

```python
overall_fig = plt.figure()
for path in paths[4:]:
    name, dataset, _ = path.split('/')
    df = pd.read_csv(path)
    acc_before = df['accuracy_test_before'].values
    acc_after = df['accuracy_test_after'].values
    epochs = df['epoch'].values
    
    plt.plot(epochs, acc_before, label=f'{name}')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.title(f'Overall_{dataset}')
    plt.legend()
    overall_fig.savefig(f'Overall_{dataset}_before.png')
```
