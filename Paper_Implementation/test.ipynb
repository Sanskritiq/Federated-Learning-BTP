# Dirilitch's Theorem

```python
import numpy as np
import matplotlib.pyplot as plt
s = np.random.dirichlet((10, 10, 10, 10), 20).transpose()
plt.barh(range(20), s[0])
plt.barh(range(20), s[1], left=s[0], color='g')
plt.barh(range(20), s[2], left=s[0]+s[1], color='r')
plt.barh(range(20), s[3], left=s[0]+s[1]+s[2], color='b')
plt.title("Lengths of Strings")
```

```python
s = np.random.default_rng().dirichlet((0.5, 0.5, 0.5), 20).transpose()
import matplotlib.pyplot as plt
plt.barh(range(20), s[0])
plt.barh(range(20), s[1], left=s[0], color='g')
plt.barh(range(20), s[2], left=s[0]+s[1], color='r')
plt.title("Lengths of Strings")
```
